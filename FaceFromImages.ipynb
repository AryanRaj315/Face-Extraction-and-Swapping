```python
import cv2
import os
```

# Using cv2's Cascade Classifier to detect faces in images and saving them

```python
def getFace(path, location):
    try:
        image = cv2.imread(path)
        rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    except:
        return
    faceCascade = cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_frontalface_default.xml")
    faces = faceCascade.detectMultiScale(
            rgb,
            scaleFactor=1.3,
            minNeighbors=3,
            minSize=(30, 30)
    ) 
    print("Found {0} Faces!".format(len(faces)))
    for (x, y, w, h) in faces:
        cv2.rectangle(image, (x, y), (x+w, y+h), (0, 255, 0), 2)
        roi_color = image[y:y + h, x:x + w] 
        print("Saving.....") 
        cv2.imwrite(location+str(w) + str(h) + '_faces.jpg', roi_color)
```

Get all of the faces from the images stored in the directory saved to the given location 

```python
def ExtractImages(directory, location):
    for filename in os.listdir(directory):
        if filename.endswith(".jpg"): 
            path = (os.path.join(directory, filename))
            getFace(path, location)
        else:
            continue
```
